–ó–∞–¥–∞—á–∞ 1. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –æ–¥–∏–Ω –∞—Ä–≥—É–º–µ–Ω—Ç ùëõ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å—É–º–º—É
–≤—Å–µ—Ö –Ω–∞—Ç—É—Ä–∞–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, –º–µ–Ω—å—à–∏—Ö ùëõ –∏ —Ç–∞–∫–∏—Ö, –∫–æ—Ç–æ—Ä—ã–µ –¥–µ–ª—è—Ç—Å—è –Ω–∞ 3 –∏–ª–∏ –Ω–∞ 5. –ü—Ä–∏–º–µ—Ä:

```python
def sumThreeFive(n):
    s = 0
    for i in range(1, n):
        
        if ((i % 3 == 0)  or (i % 5 == 0) ):
            
            s +=i
    return (s)
```

—á–ª–µ–Ω –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –§–∏–±–æ–Ω–∞—á—á–∏

```python
def fibo (n):
    s = 0
    if n == 1:
        s = 1
    if n ==2 :
        s = 2
    if n> 2:
        s = fibo(n-1)+ fibo(n-2)
    return(s)
```

```python
#–¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏
for i in range( 1, 11):
    print(fibo(i))
```

```python
#–≤—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è, –∫–æ—Ç–æ—Ä–∞—è –∫–∞–∂–¥–æ–º—É —Å–ª–æ–≤—É —Å–æ–ø–æ—Å—Ç–æ–≤–ª—è–µ—Ç 1 –∏–ª–∏ 0, –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ç–æ–≥–æ, –º–æ–∂–Ω–æ –ª–∏ –µ–≥–æ —Å–æ—Å—Ç–∞–≤–∏—Ç—å –ø–æ —É—Å–ª–æ–≤—é –∑–∞–¥–∞—á–∏
def words (word):
    answer = 1
    for l in  word:

        if (( '–ª–µ–∫–∞—Ä—Å—Ç–≤–æ'.count(l ) ) < word.count(l )) :
            answer = 0
    return (answer)
```

```python
def typesetter(file, word ):
    data = pd.read_csv( file, header=None)
    to_print = list(data [ data.apply(  lambda x :  words(x[0])  , axis = 1)>0][0])
    for i in range(len(to_print) ):
        print ( to_print[i])
```

```python
def  guess_the_word_leading ( file  )  :
    data = pd.read_csv(file , header=None)
    data['len_is_5'] = data.apply ( lambda x : (len(x[0]) ==5)*1 ,axis =1 ) 
    data5 = data[data['len_is_5'] == 1]   
    
    good_secret_word_is_find = 0
    secret_word = ''
    
    #–≤—Å–µ –±—É–∫–≤—ã –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ä–∞–∑–Ω—ã–º–∏
    while good_secret_word_is_find  ==0 :
        secret_word = (list(data5[0]) [ random.randint(0, len(data5)) ])
        cond =0
        for c in secret_word:
            if secret_word.count(c) > 1:
                cond = 1
            
        
        if  cond == 0   :
            good_secret_word_is_find = 1
    
    is_guess = 0 
    N = 0
    print ('( –¥–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏)secret_word '+ secret_word)  #–≤—ã–≤–æ–¥ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ —Ä–∞–∑—Ä–∞–±–æ—Ç–∫–∏
    while is_guess == 0:
        right_letters = 0
        N +=1
        word = input("–í–≤–µ–¥–∏—Ç–µ —Å–ª–æ–≤–æ: ")
        if len(word) != 5 :
            print ('–°–ª–æ–≤–æ –¥–æ–ª–∂–Ω–æ —Å–æ–¥–µ—Ä–∂–∞—Ç—å 5 –±—É–∫–≤')
        else:
            if word == secret_word :
                print ('–£–≥–∞–¥–∞–ª–∏ —Å –ø–æ–ø—ã—Ç–∫–∏ –Ω–æ–º–µ—Ä '+str(N))
                is_guess = 1
            else:
                #–ø–æ–¥—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–≥–∞–¥–∞–Ω–Ω—ã—Ö –±—É–∫–≤
                for letter in word:
                    right_letters += (secret_word.count( letter) >0)*1
                print (str(right_letters))
```

```python
def all_letters_is_uniq_func(word):
    all_letters_is_uniq = 1 
    for c in word :
        if word.count(c) >1 :
            all_letters_is_uniq =0 
            break
    if all_letters_is_uniq ==1 :
        return (1)
    else:
        return (0)
```
